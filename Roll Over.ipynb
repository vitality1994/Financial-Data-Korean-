### Roll over in Future

Traders roll over futures contracts to switch from the front month contract 

that is close to expiration to another contract in a further-out month.

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel(r'C:\Users\jylee\Desktop\Financial_Engineering_Recipe_Excel\11.2 WTI.xlsx', 'Spot', index_col = 'Date')  
df2 = pd.read_excel(r'C:\Users\jylee\Desktop\Financial_Engineering_Recipe_Excel\11.2 WTI.xlsx', 'Front', index_col = 'Date')  
df3 = pd.read_excel(r'C:\Users\jylee\Desktop\Financial_Engineering_Recipe_Excel\11.2 WTI.xlsx', 'Second', index_col = 'Date')  
```

```python
df = pd.concat([df1, df2, df3], axis = 1)
```

```python
df = df.sort_index(axis = 0)
```

```python
df = df.fillna(method = 'ffill')
df = df.fillna(method = 'bfill')
```

```python
df
```

```python
df_e = pd.read_excel(r'C:\Users\jylee\Desktop\Financial_Engineering_Recipe_Excel\11.2 WTI.xlsx', 'Expiry')
df_e
```

```python
expiry = df_e['Last Trade']
# expiry is the Series includes one column.

print(expiry.values)
print(pd.to_datetime(expiry.values))
# if you don't use the function 'pd.to_datetime' we cannot compare between dates of Last Price
# and the index of df because they have different form of
```

```python
df['Expiry']  = np.nan
df
```

```python
for d in df.index:
    if d in pd.to_datetime(expiry.values):
        df.loc[d].Expiry = 1

df[-10:-1]

# if there is 1 in Expiry column, that means it is the expiration date of the future.
```

```python
df
```

```python
df.index
```

```python
price_before = df.loc['2010-01-04', 'Spot']
price_now = df.loc['2017-12-28', 'Spot']
profit = (price_now - price_before) * 1000
print('Profit (price_now - price_before) : {:,.2f}'.format(profit))
```

```python
roll_over_cost_acc = 0 
for d in df.index:
    if df.loc[d].Expiry == 1 :
        roll_over_cost = df.loc[d].Second - df.loc[d].Front
        roll_over_cost_acc += roll_over_cost
        
print('Accumulated roll over cost is : {:,.2f}'.format(roll_over_cost_acc))
roll_over_cost_acc *= 1000
```

```python
profit
```

```python
roll_over_cost_acc
```

```python
final_profit = profit - roll_over_cost_acc
print('Accumulated profit including roll over cost is : {:,.2f}'.format(final_profit))
```
