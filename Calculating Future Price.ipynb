```python
# Theoretical prices of Future
# F = S*e^((r-d)*t)
```

```python
# In Financial Engineering, One year is regarded as about 250 days. 
# The days left until the expiration date  -> (last contraction date - base date) / 252
```

```python
def future_price(S, r, d, T, t0):
    t = np.busday_count(t0.date(), T.date()) / 252
# What I have to remember here is that when I want to use a function for counting days between
# two different dates, I have to left only date part of dates, which means I have to eliminate 
# hour, minute, second part.

    F = S*np.exp((r-d)*t)
    return F
# function date() gives only date part.(without hour, minute, second)
```

```python
T = pd.to_datetime('2017.12.14')
r = 0.0225 # interest rate
d = 0.015 # dividend rate
for n in df.index:
    t0 = n
    S = df.loc[n,'Index']
    F = future_price(S, r, d, T, t0)
    f = '{:,.2f}'.format(F) 
    # express numbers to 2 decimal places using the formating method.
    df.loc[n, 'Theoretical Prices 1'] = f
    # n is already date type and it goes to the index.
    # '이론가1' is the name of the new column that I want to make.
    # and f is prices of the future.
```

```python
# pricing of a future using method of simple interest rate.
# F = S[1 + (r-d)*(t/365)]

def S_future_price(S, r, d, T, t0):
    t = (T - t0).days / 365  
    

    
    F = S * (1 + (r - d)*t)
    
    return F
```

```python
T = pd.to_datetime('2017.12.14')
r = 0.0225
d = 0.015
for n in df.index:
    t0 = n
    S = df.loc[n,'Index']
    F = S_future_price(S, r, d, T, t0)
    f = '{:,.2f}'.format(F) 
    df.loc[n, 'Theoretical Prices2'] = f
```
